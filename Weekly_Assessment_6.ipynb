```python
# 1. Write a recursive function for the Fibonacci sequence
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)
terms = int(input("Enter the number of terms: "))
print("Fibonacci Sequence:")
for i in range(terms):
    print(fibonacci(i), end=" ")
```

```python
# 2. Explain "Memoization" and how it helps recursive functions.
# Memoization stores the results of previous function calls
# so the same calculations are not repeated.
from functools import lru_cache
@lru_cache(maxsize=None)
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)
num = 10
print(f"Fibonacci({num}) =", fibonacci(num))
```

```python
# 3. Use reduce() from functools to find the product of a list.
from functools import reduce
numbers = [2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print("Product =", product)
```

```python
# 4. Write a function to check if a number is an Armstrong Number.
def is_armstrong(num):
    digits = str(num)
    power = len(digits)
    total = sum(int(digit) ** power for digit in digits)
    return total == num
number = int(input("Enter a number: "))
if is_armstrong(number):
    print(number, "is an Armstrong Number.")
else:
    print(number, "is not an Armstrong Number.")
```

```python
# 5. Create a closure that remembers a base value and adds it to an input.
def create_adder(base):
    def add(number):
        return base + number
    return add
add10 = create_adder(10)
print(add10(5))
print(add10(20))
```

```python
# 6. Explain the LEGB rule in Python scope.
# LEGB stands for:
# L - Local
# E - Enclosing
# G - Global
# B - Built-in
x = "Global"
def outer():
    x = "Enclosing"
    def inner():
        x = "Local"
        print("Inside inner:", x)
    inner()
    print("Inside outer:", x)
outer()
print("Outside:", x)
```

```python
# 7. Write a function that converts a decimal number to binary without bin().
def decimal_to_binary(n):
    if n == 0:
        return "0"
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n = n // 2
    return binary
number = int(input("Enter a decimal number: "))
print("Binary =", decimal_to_binary(number))
```

```python
# 8. Create a Generator function using yield to produce infinite even numbers.
def even_numbers():
    num = 0
    while True:
        yield num
        num += 2
generator = even_numbers()
for i in range(10):
    print(next(generator))
```

```python
# 9. How do you pass a function as an argument to another function?
def square(x):
    return x * x
def calculate(func, value):
    return func(value)
print(calculate(square, 5))
```

```python
# 10. Write a function to validate a password
# (Minimum 8 characters, at least 1 digit and 1 symbol)
import string
def validate_password(password):
    if len(password) < 8:
        return False
    has_digit = any(char.isdigit() for char in password)
    has_symbol = any(char in string.punctuation for char in password)
    return has_digit and has_symbol
password = input("Enter password: ")
if validate_password(password):
    print("Valid Password")
else:
    print("Invalid Password")
```
