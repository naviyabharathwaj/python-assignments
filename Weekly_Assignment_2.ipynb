```python
# 1. Join a list of strings into a single sentence
words = ["Python", "is", "a", "programming", "language"]
sentence = " ".join(words)
print(sentence)
```

```python
# 2. Remove duplicates from a list using a set
numbers = [1, 2, 3, 2, 4, 5, 3, 1]
unique_numbers = list(set(numbers))
print(unique_numbers)
```

```python
# 3. Check if a specific key exists in a dictionary
student = {
    "name": "Naviya",
    "age": 18,
    "course": "Artificial Intelligence"
}
if "age" in student:
    print("Key exists")
else:
    print("Key does not exist")
```

```python
# 4. Difference between is and ==
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)
print(a is b)
```

Explanation:
== compares the values of objects.
is checks whether both variables refer to the same object in memory.

```python
# 5. Simple calculator using if-elif-else
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")
if operator == "+":
    print("Result =", num1 + num2)
elif operator == "-":
    print("Result =", num1 - num2)
elif operator == "*":
    print("Result =", num1 * num2)
elif operator == "/":
    if num2 != 0:
        print("Result =", num1 / num2)
    else:
        print("Cannot divide by zero")
else:
    print("Invalid operator")
```

```python
# 6. Find the largest number in a list without using max()
numbers = [12, 45, 7, 89, 23]
largest = numbers[0]
for num in numbers:
    if num > largest:
        largest = num
print("Largest number:", largest)
```

```python
# 7. Use the .strip() method on a messy input string
text = "     My name is Naviya     "
clean_text = text.strip()
print("Original Text:", repr(text))
print("Stripped Text :", repr(clean_text))
```

```python
# 8. Convert a float to an integer
num = 12.89
integer_num = int(num)
print(integer_num)
```

```python
# 9. Print the multiplication table of a given number
num = int(input("Enter a number: "))
for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

```python
# 10. Use of the in keyword in Python
fruits = ["apple", "banana", "orange"]
if "banana" in fruits:
    print("Banana is present")
else:
    print("Banana is not present")
```
