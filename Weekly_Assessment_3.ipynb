```python
# 1. while loop that terminates when a user types "exit"
while True:
    user_input = input("Enter something (type 'exit' to quit): ")
    if user_input.lower() == "exit":
        print("Program terminated.")
        break
    print("You entered:", user_input)
```

```python
# 2. Dictionary of 5 students and their marks; print only those who passed
students = {
    "Alice": 85,
    "Bob": 30,
    "Charlie": 67,
    "David": 40,
    "Emma": 25
}
print("Students who passed:")
for student, marks in students.items():
    if marks > 35:
        print(student, ":", marks)
```

```python
# 3. List comprehension to square all numbers in a list.
numbers = [1, 2, 3, 4, 5]
squares = [num**2 for num in numbers]
print("Squared list:", squares)
```

```python
# 4. Merging two dictionaries in Python 3.9+
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
merged_dict = dict1 | dict2
print("Merged Dictionary:")
print(merged_dict)
```

```python
# 5. Finding the frequency of each character in a string.
text = input("Enter a string: ")
frequency = {}
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1
print("Character frequencies:")
print(frequency)
```

```python
# 6. enumerate() function in a for loop.
fruits = ["Apple", "Banana", "Orange", "Mango"]
for index, fruit in enumerate(fruits):
    print(index, "-", fruit)
```

```python
# 7. Program to find the intersection of two sets.
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
intersection_set = set1.intersection(set2)
print("Intersection of the sets:")
print(intersection_set)
```

```python
# 8. zip() to combine two lists into a dictionary.
keys = ["name", "age", "city"]
values = ["Naviya", 19, "Chennai"]
person = dict(zip(keys, values))
print(person)
```

```python
# 9. 3x3 matrix using nested lists and access the middle element.
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print("Matrix:")
for row in matrix:
    print(row)
middle_element = matrix[1][1]
print("Middle element:", middle_element)
```

```python
# 10. Script to sort a dictionary by its values.
students = {
    "Alice": 85,
    "Bob": 70,
    "Charlie": 92,
    "David": 65
}
sorted_dict = dict(sorted(students.items(), key=lambda item: item[1]))
print("Dictionary sorted by values:")
print(sorted_dict)
```
