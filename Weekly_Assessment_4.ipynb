```python
# 1. Explain the break, continue, and pass statements with examples
# break: Exits the loop immediately
print("Example of break:")
for i in range(1, 6):
    if i == 3:
        break
    print(i)
# continue: Skips the current iteration
print("\nExample of continue:")
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
# pass: Placeholder that does nothing
print("\nExample of pass:")
for i in range(1, 6):
    if i == 3:
        pass
    print(i)
```

```python
# 2. Convert a list of temperatures in Celsius to Fahrenheit using map()
celsius = [0, 20, 37, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

```python
# 3. Find the second smallest element in a list
numbers = [12, 5, 8, 3, 10, 3, 7]
unique_numbers = sorted(set(numbers))
if len(unique_numbers) >= 2:
    print("Second smallest element:", unique_numbers[1])
else:
    print("No second smallest element.")
```

```python
# 4. Write a program to check if a year is a leap year
year = int(input("Enter a year: "))
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

```python
# 5. Reverse a dictionary (keys become values and vice versa)
student = {
    "Alice": 85,
    "Bob": 90,
    "Charlie": 78
}
reversed_dict = {value: key for key, value in student.items()}
print("Original Dictionary:", student)
print("Reversed Dictionary:", reversed_dict)
```

```python
# 6. Demonstrate the use of any() and all()
bool_list = [True, True, False, True]
print("List:", bool_list)
print("any():", any(bool_list))
print("all():", all(bool_list))
```

```python
# 7. Flatten a nested list
nested_list = [[1, 2], [3, 4], [5, 6]]
flat_list = [item for sublist in nested_list for item in sublist]
print("Flattened List:", flat_list)
```

```python
# 8. Difference between append() and extend()
list1 = [1, 2, 3]
# append()
list2 = list1.copy()
list2.append([4, 5])
print("Using append():", list2)
# extend()
list3 = list1.copy()
list3.extend([4, 5])
print("Using extend():", list3)
# append() adds the entire list as one element.
# extend() adds each element individually.
```

```python
# 9. Find the longest word in a sentence
sentence = input("Enter a sentence: ")
words = sentence.split()
longest_word = max(words, key=len)
print("Longest word:", longest_word)
```

```python
# 10. Find the 3 most common items using Counter
from collections import Counter
items = ['apple', 'banana', 'apple', 'orange', 'banana',
         'apple', 'grape', 'banana', 'apple']
counter = Counter(items)
print("3 Most Common Items:")
print(counter.most_common(3))
```
