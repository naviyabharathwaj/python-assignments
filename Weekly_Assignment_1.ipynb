1.
**List**: Mutable (changeable), slower, and consumes more memory because it allocates extra space to accommodate future appends/extensions.

**Tuple**: Immutable (unchangeable), faster, and consumes less memory because its size is fixed at creation.

```python
my_list = [1, 2, 3, 4, 5]
my_tuple = (1, 2, 3, 4, 5)

print(f"List memory usage: {len(my_list)} bytes")
print(f"Tuple memory usage: {len(my_tuple)} bytes")
```

2. Python uses tuple unpacking to swap variables cleanly without needing a temporary third variable.

```python
a = 10
b = 20

a, b = b, a

print(f"a: {a}, b: {b}")
```

3. Find the Data Type of Every Element in a Mixed List

```python
mixed_list = [42, "Hello", 3.14, True, [1, 2], {"key": "value"}]

for element in mixed_list:
    print(f"Element: {element} | Type: {type(element)}")
```

4. Check if a String is a Palindrome

```python
def is_palindrome(text):
    clean_text = text.replace(" ", "").lower()
    return clean_text == clean_text[::-1]

test_str = "Radar"
print(f"Is '{test_str}' a palindrome?: {is_palindrome(test_str)}")
```

5. / is Float Division: Always returns a decimal (float), even if the numbers divide evenly.
// is Floor Division: Divides the numbers and rounds down to the nearest whole integer.

```python
print(f"Float Division (7 / 2): {7 / 2}")
print(f"Floor Division (7 // 2): {7 // 2}")
```

6. Negative indexing allows us to extract the last N characters of a string.

```python
text = "Artificial Intelligence"
last_three = text[-3:]
print(f"Last 3 characters: {last_three}")
```

7. Count the Number of Vowels in a Sentence

```python
sentence = input("Enter a sentence: ")
vowels = "aeiouAEIOU"
vowel_count = sum(1 for char in sentence if char in vowels)
print(f"Number of vowels: {vowel_count}")
```

8. **f-strings** (formatted string literals) allows us to embed variables or expressions directly inside a string using curly braces {}. It is cleaner, faster, and more readable than older formatting methods like % or .format().

```python
name = "Naviya"
age = 18

print(f"My name is {name} and I am {age} years old.")

print(f"Next year, I will be {age + 1} years old.")
```

9. Print Even Numbers from 1 to 20

```python
numbers = list(range(1, 21))
even_numbers = [num for num in numbers if num % 2 == 0]
print(even_numbers)
```
